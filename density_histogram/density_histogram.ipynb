We load data on monthly stock returns from 1926 to 2021 ([source](https://mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html)).

```python
import numpy as np
import pandas as pd

stocks = pd.read_csv("stocks.csv")
ret = stocks["Mkt-RF"].values
```

A _density histogram_ is a histogram that is normalized so that the total area under the bars sums to unity. Argue that the stock return is not normally distributed as follows: 

- Plot a density histogram of the returns $\texttt{ret}$ with 100 bins. 
- Over this histogram, plot the PDF of a normal distribution with mean equal to the sample mean of the returns and standard deviation equal to the sample standard deviation of the returns.

```python
import seaborn as sns
import matplotlib.pyplot as plt


# plotting histogram for carat using distplot()
sns.distplot(a=ret, kde=True, bins=100)
# visualizing plot using matplotlib.pyplot library
plt.savefig('density_histogram.pdf', bbox_inches='tight')
plt.show()
```

```python
import math


sns.distplot(a=ret, kde=True, bins=100)
mu,sigma= ret.mean(),math.sqrt(ret.var())
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 500)
y = np.exp(-(x - mu) ** 2 / (2 * sigma ** 2)) / (math.sqrt(2*math.pi)*sigma)
plt.plot(x, y, "g", linewidth=2)
plt.grid(True)
# visualizing plot using matplotlib.pyplot library
plt.savefig ('add_normal_distribution.pdf', bbox_inches='tight')
plt.show()
```
