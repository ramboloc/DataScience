It can be problematic to use standard kernel density estimation to approximate a density with bounded support. Demonstrate this as follows:

- Simulate lognormal data.
- Fit Gaussian kernel density estimates using `scipy.stats.gaussian_kde`.
- Display three plots showing the tradeoff between fit at the support boundary and overall fit.

Explain why the phenomenon demonstrated in the plots occurs.

Simulate lognormal data.

### Simulate lognormal data.


```python
from scipy.stats.kde import gaussian_kde
from scipy.stats import lognorm
from numpy import hstack

sample1 = lognorm.rvs(s=0.6,loc=0, scale=1, size=500)
sample3 = lognorm.rvs(s=0.6,loc=0, scale=1, size=10000)
sample1
```

### Fit Gaussian kernel density estimates using scipy.stats.gaussian_kde


```python
s=0.6
sample = hstack([sample1])
sample3 = hstack([sample3])
probDensityFun = gaussian_kde(sample,bw_method='scott')
probDensityFun3 = gaussian_kde(sample3,bw_method='scott')
```

### Display three plots showing the tradeoff between fit at the support boundary and overall fit.

```python
import numpy as np
from matplotlib import pyplot as plt


plt.title("KDE Demonstration with sample 500", fontsize=20)
plt.hist(sample, bins = 10, color='purple', edgecolor='black',density=True)
x = np.linspace(lognorm.ppf(0.001, s), lognorm.ppf(0.999, s), 1000)
plt.plot(x, probDensityFun(x), 'r')
plt.grid(True)
# visualizing plot using matplotlib.pyplot library
plt.savefig ('Gaussian_kernel_density_estimates500_10bins.pdf', bbox_inches='tight')
plt.show()


plt.title("KDE Demonstration with sample 500 and 100 bins", fontsize=20)
plt.hist(sample, bins = 100, color='purple', edgecolor='black',density=True)
x = np.linspace(lognorm.ppf(0.001, s), lognorm.ppf(0.999, s), 1000)
plt.plot(x, probDensityFun(x), 'r')
plt.grid(True)
# visualizing plot using matplotlib.pyplot library
plt.savefig ('Gaussian_kernel_density_estimates500_100bins.pdf', bbox_inches='tight')
plt.show()

plt.title("KDE Demonstration with sample 10000 and 30 bins", fontsize=20)
plt.hist(sample, bins = 30, color='purple', edgecolor='black',density=True)
x = np.linspace(lognorm.ppf(0.001, s), lognorm.ppf(0.999, s), 5000)
plt.plot(x, probDensityFun3(x), 'r')
plt.grid(True)
# visualizing plot using matplotlib.pyplot library
plt.savefig ('Gaussian_kernel_density_estimates10000_30bins.pdf', bbox_inches='tight')
plt.show()
```

### Explanation



The histogram reflects the distribution of real samples. The red curve is the result of Gaussian kernel density estimation. When the sample size is 500, we use 10 bins, and we find that the Gaussian kernel density estimation is not smooth. The guess may be related to the sample size and the number of bins. For this reason, we changed the sample and bins to get better fitting results when the number of samples is 10000 and the bins is 30. Smaller samples cannot well represent the original distribution of variables, and too few or too many Bins cannot represent the change of variable distribution.
